### 데이터 전처리 - 파일 양식 통일

```python
#대륙이름 삭제하기
def drop_conti(df):
    conti_list = ['아시아', '북아메리카', '남아메리카', '유럽', '오세아니아']

    if '국가' in df.columns:
        df = df[~df['국가'].isin(conti_list)]
        df=df.rename(columns={'국가' : '국가별'})
    elif '국가별' in df.columns:
        df = df[~df['국가별'].isin(conti_list)]

    return df


PersonalIncome_df=drop_conti(PersonalIncome_df)
EcoGrowth_df=drop_conti(EcoGrowth_df)
ttlIncome_df=drop_conti(ttlIncome_df)
ConsumeIndex_df=drop_conti(ConsumeIndex_df)
FutureInpl_df=drop_conti(FutureInpl_df)
Debt_df=drop_conti(Debt_df)
Debt_df.head()
```

```python
#소비자물가지수 컬럼 변경
ConsumeIndex_df.iloc[0, 1:] = ConsumeIndex_df.iloc[0, 1:].astype(str).str.split('.').str[0]
ConsumeIndex_df.iloc[0, 1:] = ConsumeIndex_df.iloc[0, 1:].apply(pd.to_numeric, errors='coerce')
ConsumeIndex_df.iloc[0, 1:] = ConsumeIndex_df.iloc[0, 1:].astype('int32')
ConsumeIndex_df.columns = ConsumeIndex_df.iloc[0]
ConsumeIndex_df = ConsumeIndex_df[1:]
ConsumeIndex_df.head()
```

### 데이터 전처리 - 피벗

```python
#행열변경 테스트(피벗)
df2 = PersonalIncome_df.melt(id_vars='국가별', var_name='연도', value_name='1인당국민총소득')
df2['연도'] = pd.to_numeric(df2['연도'], errors='coerce')
df2
```

```python
def pivot(df, nation, year, value):
    df = df.melt(id_vars=[nation], var_name=year, value_name=value)
    df[year] = pd.to_numeric(df[year], errors='coerce')
    return df

PersonalIncome_pivot = pivot(PersonalIncome_df, '국가별', '연도별', '1인당국민총소득(달러)')
EcoGrowth_pivot = pivot(EcoGrowth_df, '국가별', '연도별', '경제성장률(%)')
ttlIncome_pivot = pivot(ttlIncome_df, '국가별', '연도별', '국민총소득(10억달러)')
ConsumeIndex_pivot = pivot(ConsumeIndex_df, '국가별', '연도별', '소비자물가지수(2010=100)')
FutureInpl_pivot = pivot(FutureInpl_df, '국가별', '연도별', '인플레이션전망(%)')
Debt_pivot = pivot(Debt_df, '국가별', '연도별', 'GDP대비국가부채(%)')
Debt_pivot.head()
```

```python
no_nan_val=test.iloc[0, test.columns.get_loc(no_nan_col[0])]
next_val=test.iloc[0, test.columns.get_loc(no_nan_col[0]) + 1]
cal_val=(next_val-no_nan_val)/100
test.iloc[0, test.columns.get_loc(no_nan_col[0]) - 1]=no_nan_val-(no_nan_val/100)
fill_na=test.iloc[0, test.columns.get_loc(no_nan_col[0]) - 1]
print('채울 결측값 :',fill_na)
print('nan 아닌 값 :', no_nan_val)
print('nan 아닌 다음값 :',next_val)
print('계산율:',cal_val)
```

```python
for idx, col in test.iterrows():
  if test.iloc[idx].isnull().any():
    no_nan_col = test.iloc[:, 1:].apply(lambda x: x.first_valid_index(), axis=1) #결측치없는 첫번째 컬럼 찾기
    for n in range(len(no_nan_col)):
      no_nan_val = test.iloc[idx, test.columns.get_loc(no_nan_col[n])] #결측치없는 첫번째 컬럼의 값

      for col in range(test.columns.get_loc(no_nan_col[0]) - 1, -1, -1):
        prev_val = test.iloc[0, col]
        if not pd.notna(prev_val):
            test.iloc[0, col] = no_nan_val - (no_nan_val / 100)
            no_nan_val = test.iloc[0, col]
  else :
    continue
```
