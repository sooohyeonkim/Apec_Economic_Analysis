### 데이터 전처리 - 파일 양식 통일

```python
#대륙이름 삭제하기
def drop_conti(df):
    conti_list = ['아시아', '북아메리카', '남아메리카', '유럽', '오세아니아']

    if '국가' in df.columns:
        df = df[~df['국가'].isin(conti_list)]
        df = df.rename(columns={'국가': '국가별'})
        df['국가별'] = df['국가별'].str.strip()
    elif '국가별' in df.columns:
        df = df[~df['국가별'].isin(conti_list)]
        df['국가별'] = df['국가별'].str.strip()
    else:
        pass

    return df


PersonalGNI_df=drop_conti(PersonalGNI_df)
GNI_df=drop_conti(GNI_df)
EcoGrowth_df=drop_conti(EcoGrowth_df)
ConsumeIndex_df=drop_conti(ConsumeIndex_df)
Debt_df=drop_conti(Debt_df)
Debt_df.head()
```

#### 피벗 적용

```python
def pivot(df, nation, year, value):
    df = df.melt(id_vars=[nation], var_name=year, value_name=value)
    df[year] = pd.to_numeric(df[year], errors='coerce')
    df[value] = pd.to_numeric(df[value], errors='coerce')
    return df

EcoGrowth_pivot = pivot(EcoGrowth_df, '국가별', '연도별', '경제성장률(GDP,%)')
Debt_pivot = pivot(Debt_df, '국가별', '연도별', 'GDP대비국가부채(%)')
GNI_pivot = pivot(GNI_df, '국가별', '연도별', '국민총소득(GNI, 달러)')
PersonalGNI_pivot = pivot(PersonalGNI_df, '국가별', '연도별', '1인당국민총소득(GNI, 달러)')
ConsumeIndex_pivot = pivot(ConsumeIndex_df, '국가별', '연도별', '소비자물가지수(2010=100)')

Debt_pivot.head()
```

### 데이터 전처리 - merge

```python
def From1989(df) :
  for n in range (1961,1989) :
    idx = df[df['연도별'] == n].index #apec 출범 이후 (1989 이후로 기간정리)
    df.drop(idx , inplace=True)
    df.reset_index(drop=True, inplace=True) #인덱스 통일시킴
  return df

EcoGrowth_from1989 = From1989(EcoGrowth_pivot)
Debt_from1989 = From1989(Debt_pivot)
GNI_from1989 = From1989(GNI_pivot)
PersonalGNI_from1989 = From1989(PersonalGNI_pivot)
ConsumeIndex_from1989 = From1989(ConsumeIndex_pivot)
print(Debt_from1989.head())
print(EcoGrowth_from1989.head())
```

```python
merged = pd.merge(Debt_from1989,EcoGrowth_from1989, how='left', on=['국가별', '연도별'])
merged
```

```python
merged = pd.merge(GNI_from1989,PersonalGNI_from1989, how='left', on=['국가별', '연도별'])
merged = pd.merge(merged, EcoGrowth_from1989, how='left', on=['국가별', '연도별'])
merged = pd.merge(merged, Debt_from1989, how='left', on=['국가별', '연도별'])
merged = pd.merge(merged, ConsumeIndex_from1989, how='left', on=['국가별', '연도별'])
merged
```
