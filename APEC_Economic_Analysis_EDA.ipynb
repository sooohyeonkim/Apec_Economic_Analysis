### 데이터 전처리 - 파일 양식 통일

```python
#대륙이름 삭제하기
def drop_conti(df):
    conti_list = ['아시아', '북아메리카', '남아메리카', '유럽', '오세아니아']

    if '국가' in df.columns:
        df = df[~df['국가'].isin(conti_list)]
        df = df.rename(columns={'국가': '국가별'})
        df['국가별'] = df['국가별'].str.strip()
    elif '국가별' in df.columns:
        df = df[~df['국가별'].isin(conti_list)]
        df['국가별'] = df['국가별'].str.strip()
    else:
        pass

    return df


PersonalGNI_df=drop_conti(PersonalGNI_df)
GNI_df=drop_conti(GNI_df)
EcoGrowth_df=drop_conti(EcoGrowth_df)
ConsumeIndex_df=drop_conti(ConsumeIndex_df)
Debt_df=drop_conti(Debt_df)
Debt_df.head()
```

#### 피벗 적용

```python
def pivot(df, nation, year, value):
    df = df.melt(id_vars=[nation], var_name=year, value_name=value)
    df[year] = pd.to_numeric(df[year], errors='coerce')
    df[value] = pd.to_numeric(df[value], errors='coerce')
    return df

EcoGrowth_pivot = pivot(EcoGrowth_df, '국가별', '연도별', '경제성장률(GDP,%)')
Debt_pivot = pivot(Debt_df, '국가별', '연도별', 'GDP대비국가부채(%)')
GNI_pivot = pivot(GNI_df, '국가별', '연도별', '국민총소득(GNI, 달러)')
PersonalGNI_pivot = pivot(PersonalGNI_df, '국가별', '연도별', '1인당국민총소득(GNI, 달러)')
ConsumeIndex_pivot = pivot(ConsumeIndex_df, '국가별', '연도별', '소비자물가지수(2010=100)')

Debt_pivot.head()
```

```python
def calculate_and_update(df, row_index, year_1, year_2, target_year):
    df.loc[row_index, target_year] = (df.loc[row_index, year_1]+df.loc[row_index, year_2])/2
    return df

calculate_and_update(internet_df, 18, '2001', '2005', '2003')
calculate_and_update(internet_df, 18, '2001', '2003', '2002')
calculate_and_update(internet_df, 18, '2003', '2005', '2004')

years = ['2020', '2021']
target_year = '2022'
rows_to_update = [1, 5, 7, 12, 13, 14, 15, 18, 19, 20]

for index in rows_to_update:
    calculate_and_update(internet_df, index, years[0], years[1], target_year)

internet_df
```

```python
#대륙이름 삭제하기
def drop_conti(df):
    conti_list = ['아시아', '북아메리카', '남아메리카', '유럽', '오세아니아']

    if '국가' in df.columns:
        df = df[~df['국가'].isin(conti_list)]
        df = df.rename(columns={'국가': '국가별'})
        df['국가별'] = df['국가별'].str.strip()
    elif '국가별' in df.columns:
        df = df[~df['국가별'].isin(conti_list)]
        df['국가별'] = df['국가별'].str.strip()
    else:
        pass

    return df

# Null로 바꾸고, 리셋인덱스
def ToNullandNum(df):
    df.replace('-', np.nan, inplace=True)
    df.iloc[:, 1:] = df.iloc[:, 1:].astype(float)
    df.reset_index(drop=True, inplace=True)
    return df
```

```python
mobile_df=drop_conti(mobile_df)
mobile_df.head()
```

```python
mobile_df=ToNullandNum(mobile_df)
mobile_df.head()
```

```python
mobile_df.isnull().sum()
```

```python
def pivot(df, nation, year, value):
    df = df.melt(id_vars=[nation], var_name=year, value_name=value)
    df[year] = pd.to_numeric(df[year], errors='coerce')
    df[value] = pd.to_numeric(df[value], errors='coerce')
    return df

mobile_pivot = pivot(mobile_df, '국가별', '연도별', '이동전화 가입수(인구 100명당)')
mobile_pivot.head()
```
